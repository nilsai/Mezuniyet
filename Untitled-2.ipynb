```python
df_age = df.copy()
df_age.drop("Seniority", axis=1, inplace=True)
df_age.head()
```

```python
df_seniority = df.copy()
df_seniority.drop("Age", axis=1, inplace=True)
df_seniority.head()
```

```python
age = df_age[["Age"]]
seniority = df_seniority[["Seniority"]]

df_age_ml = df_age.drop("Age", axis=1)
df_seniority_ml = df_seniority.drop("Seniority", axis=1)
```

```python
X_train_age, X_test_age, y_train_age, y_test_age = train_test_split(df_age_ml, age, test_size=0.2, random_state=42)

X_train_seniority, X_test_seniority, y_train_seniority, y_test_seniority = train_test_split(df_seniority_ml, seniority, test_size=0.2, random_state=42)
```

```python
X_train_age = X_train_age.select_dtypes(include=['int64', 'float64'])
X_test_age = X_test_age.select_dtypes(include=['int64', 'float64'])
linear_model_age = LinearRegression()
linear_model_age.fit(X_train_age, y_train_age)
y_pred_age = linear_model_age.predict(X_test_age)

mse_age = mean_squared_error(y_test_age, y_pred_age)
r2_age = r2_score(y_test_age, y_pred_age)

print("age - MSE:", mse_age)
print("age - R2:", r2_age)
```

```python
X_train_seniority = X_train_seniority.select_dtypes(include=['int64', 'float64'])
X_test_seniority = X_test_seniority.select_dtypes(include=['int64', 'float64'])
linear_model_seniority = LinearRegression()

linear_model_seniority.fit(X_train_seniority, y_train_seniority)

y_pred_seniority = linear_model_seniority.predict(X_test_seniority)

mse_seniority = mean_squared_error(y_test_seniority, y_pred_seniority)
r2_seniority = r2_score(y_test_seniority, y_pred_seniority)

print("seniority - MSE:", mse_seniority)
print("seniority - R2:", r2_seniority)
```
